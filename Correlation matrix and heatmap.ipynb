Correlation Matrix and HeapMap for iris DS

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#loading dataset

iris = sns.load_dataset('iris')
```

```python
#selecting only numeric columns from Dataset

numeric_col=iris.select_dtypes(include=["number"])
print(numeric_col.head())
```

```python
#calculating correlation matrix

corr_matrix= numeric_col.corr()
print(corr_matrix)
```

```python
#plotting heat map

plt.figure(figsize=(10,5))
sns.heatmap(corr_matrix,cmap="coolwarm", annot=True, fmt=".2f",linewidth=2)
plt.title("Heatmap for correlation matrix of IRIS dataset")
plt.show()

#cmap used to give color for heatmap
#annot=True adds the value of the box (like 1.00, -0.12,0.87.....)
#fmat=".2f" is used to set decimal values we need to show , here .2f means upto 2 decimal places
#linewidth=2 is used to seprate boxes ...giving spaces
```
